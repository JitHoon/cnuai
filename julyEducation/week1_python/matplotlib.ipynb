```python
import numpy as np
import pandas as pd

# matplotlib에서 가장 많이 사용되는 하위 모듈 matplotlib.pyplot
import matplotlib.pyplot as plt

# matplotlib 활성화
%matplotlib inline
```

## 2. matplotlib.pyplot 대표 함수들 
### (1) `plt.figure(figsize=(,))`
- 그래프 figure 사이즈 결정
### (2) `plt.plot(x, y, label="")`
- 그래프 그리는 함수
### (4) `legend()`
- 그래프 범례 (그래프 선 마킹)
### (3) `plt.show()`
- 그래프 보여주는 함수

```python
# potting을 할 figure를 선언
plt.figure(figsize=(4, 4))

# 1차 함수 (꺾은 선) 그래프를 plot
plt.plot([1, 2, 3, 4, 5], label="gragh 1")

plt.legend()

plt.show() # plt를 확인하는 명령
```

### (4) `np.arange(처음, 끝, 간격)`
- 원하는 범위와 간격을 설정하여 그래프를 그릴 수 있는 함수
### (5) `plt.title("")`
- 그래프 이름 짓기
### (6) `plt.xlabel("")`, `plt.ylabel("")`
- x, y축 이름 짓기
### (7) `plt.axis([])` -> [x_min, x_max, y_min, y_max]
- 그래프의 보고 싶은 x, y 범위 설정 가능
### (8) `plt.xticks([])`, `plt.yticks([])`
- 그래프에 눈금 만들어주기

```python
# 2차 함수 그래프 plot y = x**2 
x = np.arange(-10, 10, 0.01) # 정의역

# 그래프 이름 짓기
plt.title("y = x^2")

# 그래프 축 이름 정하기
plt.xlabel("x value")
plt.ylabel("f(x) value")

# 그래프 잘라보기
plt.axis([-5, 5, 0, 24])

# 눈금 설정
plt.xticks([i for i in range(-5, 6, 1)])
plt.yticks([i for i in range(0, 25, 3)])

# 그래프 그리기, 범례, 보여주기
plt.plot(x, x**2, label="gragh 1")
plt.legend()
plt.show()
```

## 3.  Matplotlib.pyplot의 여러 plot들 (case study)
### (1) 꺾은선 그래프 : `plt.plot()`
- 시계열 데이터에서 많이 사용된다.

```python
x = np.arange(20)
y = np.random.randint(0, 20, 20)

plt.axis([0, 20, 0, 20])

# 눈금 설정
plt.xticks([i for i in range(0, 21, 2)])
plt.yticks([i for i in range(0, 21, 5)])

plt.plot(x, y)
plt.show()
```

### (2) 산점도 : `plt.scatter(,)`
- x와 y의 관계가 별개인 경우 많이 사용된다.

```python
plt.scatter(x, y)
plt.show()
```

### (3) 박스 그림 : `plt.boxplot((,))`
- 단일 변수의 전반적인 분포를 확일할 때 경우 많이 사용된다.
- 수치형 데이터에 대한 정보를 담고있음 : 4분위수 = (Q1, Q2, Q3, min, max)
- Q1 : 25%에 해당하는 값
- Q2 : median(중앙값)에 해당하는 값
- Q3 : 75%에 해당하는 값

- 사분위수 : 데이터 테이블 또는 데이터 테이블의 일부를 거의 균등한 관찰값 수를 포함하는 4개의 그룹으로 나눈 값입니다. (ex. 합계 100%를 4개의 균등한 부분으로 분할합니다. 25%, 50%, 75% 100%.)

```python
plt.title("Box plot of x, y") # y = np.random.randint(0, 20, 20)
plt.boxplot((x, y))
plt.show()
```

### (4) 막대 그래프 : `plt.bar(,)`
- 범주형 데이터의 "값"과 그 값의 크기를 직사각형 막대로 나타낸 그림

```python
plt.bar(x, y)

plt.xticks((np.arange(0, 20, 1)))
plt.yticks((np.arange(0, 20, 1)))

plt.show()
```

### (5) 히스토그램 : `plt.hist(,)`
- `plt.bar(,)`와 동일한 그림을 보여주지만
- '계급'으로 데이터를 나타내준다. (특정 범위로 묶어서 그래프를 보여줌)
- 즉, 숫자 자료를 동일한 폭의 계급(버킷, 간격)으로 그룹화하여 자료를 나타냄.
- ex. 무게가 6~8 kg 인 호박 8개


```python
plt.hist(y, bins=np.arange(0,20,2)) # bins : 범주의 간격 설정
# y = np.random.randint(0, 20, 20)

plt.xticks(np.arange(0, 20, 2))
plt.show()
```

### (6) 원형 그래프 (Pie chart) : `plt.pie()`
- 비율을 부채꼴 모양로 나타낸 그래프

```python
z = [10, 20, 30, 40]
plt.pie(z, labels=['one','two','three','four'])
plt.show()
```

## 4. Seaborn
- Matplotlib를 기반으로 다양한 시각화 라이브러리

### (1) seaborn 시작하기

```python
import seaborn as sns
```

### (2) Kernel Density 그래프 : `sns.kdeplot(, shade)`
- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린 그림
- 단 y축을 인자의 수로 보여주지 않고 density로 보여준다.

```python
x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)
plt.xticks(np.arange(0, 21, 2))

plt.hist(y, bins=x)
plt.show()

sns.kdeplot(y, shade =True)
plt.show()
```

### (3) Count 그래프 : `sns.countplot()`
- 카운트 데이터 추출 과정

```python
# 불편한 카운트 데이터 추출 과정
vote_df = pd.DataFrame({"name":['Andy','Bob','Cat'], 'vote':[True, True, False]})
print(vote_df)

vote_count = vote_df.groupby('vote').count()
print(vote_count)

plt.bar(x=[False, True], height=vote_count['name'])
plt.show()

# 편한 카운트 데이터 추출 과정
sns.countplot(x=vote_df['vote'])
plt.show()
```

### (4) Cat plot : `sns.catplot(x = 'column name', y = 'column name', hue, data = csv 파일, kind = '그래프 종류')`
- 여러개의 그래프를 묶는 방법 (kind가 그래프 종류임)
- kind : 'violin',  

```python
covid = pd.read_csv("./country_wise_latest.csv")
covid.head(5)
```

```python
# 아래 그래프 형태는 streat 형태임
s = sns.catplot(x='WHO Region', y='Confirmed',data=covid, kind = 'violin')

# sns에서의 figure 크기 조절 방법
s.fig.set_size_inches(10,6)
plt.show()
```

- cf. kind를 따로 선언하지 않고 원하는 종류의 그래프 사용가능
- `sns.stripplot()`
- `sns.swarmplot()` : stript의 겹치는 부분을 펴서 보여줌

```python
s = sns.stripplot(x='WHO Region', y='Confirmed',data=covid)
plt.show()

s = sns.swarmplot(x='WHO Region', y='Confirmed',data=covid)
plt.show()
```

### (5) Heatmap : `sns.heatmap()`
- 데이터의 행렬을 색상으로 표현해주는 그래프

```python
# 행렬꼴로 csv파일 불러오기
covid.corr()
```
